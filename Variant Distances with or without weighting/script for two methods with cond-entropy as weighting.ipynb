````python
## count the unigram and bigram frequencies of tones and segments separately
from collections import Counter 
## if lang is None, compute across languages
def get_lang_grams(lang=None, data=segment_tone_df):
    uni_tone, uni_seg, bi_tone, bi_seg = [], [], [], []
    tones, segs = data["TOKENS_TONE"], data["TOKENS_SEG"]
    if lang:
        tones, segs = data[data["TAXA"]==lang]["TOKENS_TONE"], data[data["TAXA"]==lang]["TOKENS_SEG"]

    for tone, seg in zip(tones, segs):
        ## add the initial and end token: "-", "#"
        tone, seg = "- " + tone + " #", "- " + seg + " #"
        tone_list = tone.split()
        seg_list = seg.split()
        uni_tone += tone_list
        uni_seg += seg_list

        for i in range(len(tone_list)-1):
            bi_t = tone_list[i:i+2]
            bi_tone.append(" ".join(bi_t))
        for j in range(len(seg_list)-1):
            bi_s = seg_list[j:j+2]
            bi_seg.append(" ".join(bi_s))

    uni_tone = Counter(uni_tone)
    uni_seg = Counter(uni_seg)
    bi_tone = Counter(bi_tone)
    bi_seg = Counter(bi_seg)
    return uni_seg, uni_tone, bi_seg, bi_tone
````

````python
# total tone conditional entropy
con_entropy_t = 0
for i in uni_tone.keys():
    for j in uni_tone.keys():
        bi_count = 0
        if i+" "+j in bi_tone:
            bi_count += bi_tone[i+" "+j]
        if bi_count != 0:
            p_joint = bi_count/sum(list(bi_tone.values()))
            p_i = uni_tone[i]/sum(list(uni_tone.values()))
            
            con_entropy_t += p_joint*np.log2(bi_count/p_i)
con_entropy_t
````

````python
# total segment conditional entropy
con_entropy_s = 0
for i in uni_seg.keys():
    for j in uni_seg.keys():
        bi_count = 0
        if i+" "+j in bi_seg:
            bi_count += bi_seg[i+" "+j]
        if bi_count != 0:
            p_joint = bi_count/sum(list(bi_seg.values()))
            p_i = uni_seg[i]/sum(list(uni_seg.values()))
            
            con_entropy_s += p_joint*np.log2(bi_count/p_i)
con_entropy_s
````

Compute the conditional entropy of segment and tone in the word <br>

**x = segment or tone sequence in a concept variant** <br>
$H(x)=-\displaystyle\sum_{n=2}^{|x|}p(x_{n-1},x_n)log_2p(x_n|x_{n-1})$ 
<br>
p(x_{n-1},x_n) is defined as count of (x_{n-1},x_n)/ count of all bigrams
<!-- In practice: 
<br>
$H(x)=-\displaystyle\sum_{n=1}^{|x|}log_2p(x_n|x_{n-1})$ <br>
then normalized by the number of pairs of $(x_n, x_{n-1})$ in x. -->


````python
def compute_cond_entropy(uni_dict=uni_seg, bi_dict=bi_seg, type="segment", out=True): 
    bi_total_count = sum(list(bi_dict.values()))
    #uni_total_count = sum(list(uni_dict.values()))
    items = uni_dict.keys()
    cond_entropys = {}
    for t1 in items:
            for t2 in items:
                uni, bi = t1, t1 + " " + t2
                if bi not in cond_entropys and bi in bi_dict:
                    #print("current item:", uni)
                    #print("current and next item (bigram):", bi, "\n")
                    current_count = uni_dict[uni]
                    current_next_joint_count = bi_dict[bi]
                    #print(f"uni item'{uni}' count:",current_count)
                    #print(f"bi item '{bi}' count:", current_next_joint_count, ", bi seg total count:", bi_total_count)
                    prob_next_given_current = current_next_joint_count/current_count
                    prob_current_next_joint = current_next_joint_count/bi_total_count
                    #print(f"Probabilities:'\n'P({uni}, {t2})=P(count({bi})/total bi count):", f"{current_next_joint_count}/{bi_total_count} =", prob_current_next_joint, "\n", f"P({t2}|{t1})=P(count({bi})/count({t1})):", f"{current_next_joint_count}/{current_count} =", prob_next_given_current)
                    #print(f"Conditional Information content of the item '{t2}' given '{t1}' = -log2({prob_next_given_current}) =", -np.log2(prob_next_given_current))
                    cond_entropy = -prob_current_next_joint*np.log2(prob_next_given_current)
                    #print(f"Conditional entropy of '{bi}' = -{prob_current_next_joint}*log2({prob_next_given_current}) =", cond_entropy)
                    #print("\n")
                    #cond_entropy_seg = -np.log2(current_next_seg_joint_count/current_seg_count)
                    cond_entropys[bi] = cond_entropy
    if out:
        print(f"Total {type} conditional entropy: ", sum(cond_entropys.values()))
    return cond_entropys
## 
def get_conditional_entropy(data=segment_tone_df, out=True):
    langs = segment_tone_df["TAXA"].drop_duplicates()
    lang_cond_entropys = {}
    for lang in langs:
        uni_seg, uni_tone, bi_seg, bi_tone = get_lang_grams(lang=lang)
        if out:
            print(f"~~~~~~Working on the language *{lang}*~~~~~~")
            print("Total count of uni_tone, bi_tone, uni_seg, bi_seg:", sum(list(uni_tone.values())), sum(list(bi_tone.values())), sum(list(uni_seg.values())), sum(list(bi_seg.values())))

        seg_cond_entropys = compute_cond_entropy(uni_seg, bi_seg, out=out)
        tone_cond_entropys = compute_cond_entropy(uni_tone, bi_tone, type="tone", out=out)
        lang_cond_entropys[lang] = seg_cond_entropys, tone_cond_entropys
  

    return lang_cond_entropys
````

#### Apply the conditional entropies as weights in distance.

```
word_distance = 
segment_distance + (tone_entropy_of_the_word_pair/seg_entropy_of_word_pair)*tone_distance
```
**NOTE** the edit distance is not normalized! Since we compute segment and tone separately and add up later, but (a+b)/(c+d) != a/c + b/d.

````python
def get_weighted_distances(all_lang_pairs, tone_distance_scores, seg_distance_scores):
    lang_cond_entropys = get_conditional_entropy(out=False)
    langpair_distances = {}
    for lang_pair in all_lang_pairs:
        if lang_pair not in tone_distance_scores:
            lang_pair = lang_pair[1], lang_pair[0]
        lang1, lang2 = lang_pair[0], lang_pair[1]
        entropy1, entropy2 = lang_cond_entropys[lang1], lang_cond_entropys[lang2]      
        #TODO should we normarlize it by the total length of the two word IPA?
        # t_edits+s_edits is original edits
        tone_distances = tone_distance_scores[lang_pair]
        seg_distances = seg_distance_scores[lang_pair]
        total_tone_distance, total_seg_distance = 0, 0
        for char in tone_distances:
            total_tone_distance += tone_distances[char][2]
            total_seg_distance += seg_distances[char][2]
        # average by total number of concepts 
        avg_tone_distance = total_tone_distance/len(tone_distances.keys()) 
        avg_seg_distance = total_seg_distance/len(tone_distances.keys())  
        entropy_seg1, entropy_tone1 = sum(entropy1[0].values()), sum(entropy1[1].values()) 
        entropy_seg2, entropy_tone2 = sum(entropy2[0].values()), sum(entropy2[1].values()) 
        weight = (entropy_seg1 + entropy_seg2)/(entropy_tone1 + entropy_tone2)
        weighted_distance = weight*avg_seg_distance + avg_tone_distance
        unweighted_distance = avg_seg_distance+avg_tone_distance
        langpair_distances[lang_pair] = avg_seg_distance, avg_tone_distance, unweighted_distance, weight, weighted_distance 
        #print("Lang_pair:", lang_pair, "Weight:", weight, "Distance:", unweighted_distance, "Weighted_distance:", weighted_distance)
    return langpair_distances
````

````python
def to_file(all_distances, distance_type="1gram"):
    with open("all_langpairs_unweighted_and_weighted_distances_" + distance_type + ".tsv", "w", encoding="utf-8") as f:
        f.write("Lang_pair" + "\t" + "Segment_distance" + "\t" + "Tone_distance" + "\t" + "Unweighted_total_distance" + "\t" + "Weight" + "\t" + "Weighted_total_distance" +"\n")
        for lang_pair in all_distances:
            seg_dist, tone_dist, distance, weight, weighted_distance = all_distances[lang_pair]
            f.write(lang_pair[0] + "_" + lang_pair[1] + "\t" + str(seg_dist) + "\t" + str(tone_dist) + "\t" + str(distance) + "\t" + str(weight) + "\t" + str(weighted_distance) + "\n")
````

#### Now use the results to fill in a distance matrix for Gabmap

````python
all_langs = list(set(segment_tone_df["TAXA"]))
def get_distance_matrix(distance_results, weight=False, type="tfidf_3grams"):
    distances_df = pd.DataFrame(index=list(all_langs), columns=list(all_langs))
    for lang_pair in distance_results:
        all_distances = distance_results[lang_pair]
        lang1 = lang_pair[0]
        lang2 = lang_pair[1]
        weight_string = "weighted"
        langpair_distance = all_distances[-1] # weighted lang_pair distance
        if not weight:
            weight_string = "unweighted"
            langpair_distance = all_distances[-3] # unweighted lang_pair distance
        distances_df.at[lang1, lang2] = langpair_distance
        distances_df.at[lang2, lang1] = langpair_distance

    distances_df.fillna(int(0), inplace=True)
    distances_df.to_csv("Chinese_40dialects_distances_" + weight_string + "_" + type + ".csv", sep="\t", encoding="utf-8")
    return distances_df
````
